```python
%matplotlib inline
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import os 
import json
```

```python
homePath = os.getcwd()
print("SQA100")
df_sqa100 = pd.read_json(f'{homePath}/ChengRW100/SQAchengRW100.json') #json
display(df_sqa100)
print("QA100")
df_qa100 = pd.read_json(f'{homePath}/ChengRW100/QAchengRW100.json') #json
df_qa100 = df_qa100.sort_values('time')
display(df_qa100)

print("SQA200")
df_sqa200 = pd.read_json(f'{homePath}/ChengRW200/SQAchengRW200.json') #json
display(df_sqa200)
print("QA200")
df_qa200 = pd.read_json(f'{homePath}/ChengRW200/QAchengRW200.json') #json
df_qa200 = df_qa200.sort_values('time')
display(df_qa200)

print("SQA10000")
df_sqa10000 = pd.read_json(f'{homePath}/twitter10000/SQAtwitter10000.json') #json
display(df_sqa10000)
print("QA10000")
df_qa10000 = pd.read_json(f'{homePath}/twitter10000/QAtwitter10000.json') #json
df_qa10000 = df_qa10000.sort_values('time')
display(df_qa10000)

SMT_times = [89, 103, 318]
```

TTS = 0.99

```python
QA_TTS = [df_qa100['TTS'].min(), df_qa200['TTS'].min(), df_qa10000['TTS'].min()]
SQA_TTS = [df_sqa100['TTS'].min(), df_sqa200['TTS'].min(), df_sqa10000['TTS'].min()]
variables = [16, 58, 124]
transactions = [101, 195, 9991]
```

```python
plt.figure(figsize=(5,3))

#SQA
plt.plot(variables, SQA_TTS, label='SQA')

#QA
plt.plot(variables, QA_TTS, label='QA')

#SMT
plt.plot(variables, SMT_times, label='SMT', color="Red")

#
plt.legend(fontsize=13)
plt.xlabel("#variables",fontsize=13)
plt.ylabel("TTS(ms)",fontsize=13)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
#plt.yscale('log') #log
plt.savefig(f'{homePath}/Scaling/TTS099_scale_var_tts.pdf', bbox_inches='tight') #
plt.show()
```

```python
plt.figure(figsize=(5,3))

#SQA
plt.plot(transactions, SQA_TTS, label='SQA')

#QA
plt.plot(transactions, QA_TTS, label='QA')

#SMT
plt.plot(transactions, SMT_times, label='SMT', color="Red")

#
plt.legend(fontsize=13)
plt.xlabel("#transactions",fontsize=13)
plt.ylabel("TTS(ms)",fontsize=13)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xscale('log') #log
plt.savefig(f'{homePath}/Scaling/TTS099_scale_txn_tts.pdf', bbox_inches='tight') #
plt.show()
```

all-in-one

```python
# 
fig, axes = plt.subplots(ncols=3, figsize=(20, 4), sharey=True)
fig.subplots_adjust(wspace=0.25)


QA_TTSs = [QA_TTS, QA_TTS0999, QA_TTS09999]
SQA_TTSs = [SQA_TTS, SQA_TTS0999, SQA_TTS09999]
titles = ["r_m=0.99", "r_m=0.999", "r_m=0.9999"]


# 
for i in range(3):
    ax = axes[i]

    # 
    ax.plot(transactions, QA_TTSs[i], label="QA", color="darkorange", marker="x")
    ax.plot(transactions, SQA_TTSs[i], label="SQA", color="blue", marker="x")
    ax.plot(transactions, SMT_times, label="SMT", color="Red", marker="x")

    ax.set_title(titles[i], fontsize=13)
    ax.legend(fontsize=13)
    ax.set_xscale('log') 
    ax.tick_params(axis='x', labelsize=13)  # x
    ax.tick_params(axis='y', labelsize=13)  # y
    ax.grid('true')

# 
plt.yscale('log')
fig.text(0.5, 0.01, '#transactions', ha='center', fontsize=13)
fig.text(0.09, 0.5, 'TTS(ms)', va='center', rotation='vertical', fontsize=13)

plt.savefig(f'{homePath}/Scaling/TTS_AllInOne_scale_txn_tts.pdf', bbox_inches='tight') #

plt.show()
```
