```python
%matplotlib inline
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import os 
import json
```

```python
homePath = os.getcwd()
print("SQA100")
df_sqa100 = pd.read_json(f'{homePath}/ChengRW100/SQAchengRW100.json') #ここ変更すればjson変えられる
display(df_sqa100)
print("QA100")
df_qa100 = pd.read_json(f'{homePath}/ChengRW100/QAchengRW100.json') #ここ変更すればjson変えられる
df_qa100 = df_qa100.sort_values('time')
display(df_qa100)

print("SQA200")
df_sqa200 = pd.read_json(f'{homePath}/ChengRW200/SQAchengRW200.json') #ここ変更すればjson変えられる
display(df_sqa200)
print("QA200")
df_qa200 = pd.read_json(f'{homePath}/ChengRW200/QAchengRW200.json') #ここ変更すればjson変えられる
df_qa200 = df_qa200.sort_values('time')
display(df_qa200)

print("SQA10000")
df_sqa10000 = pd.read_json(f'{homePath}/twitter10000/SQAtwitter10000.json') #ここ変更すればjson変えられる
display(df_sqa10000)
print("QA10000")
df_qa10000 = pd.read_json(f'{homePath}/twitter10000/QAtwitter10000.json') #ここ変更すればjson変えられる
df_qa10000 = df_qa10000.sort_values('time')
display(df_qa10000)

SMT_times = [89, 103, 318]
```

TTS = 0.99の時

```python
plt.figure(figsize=(5,3))

#SQA
plt.plot(variables, SQA_TTS, label='SQA')

#QA
plt.plot(variables, QA_TTS, label='QA')

#SMT
plt.plot(variables, SMT_times, label='SMT', color="Red")

#ラベルなどの出力
plt.legend(fontsize=13)
plt.xlabel("#variables",fontsize=13)
plt.ylabel("TTS(ms)",fontsize=13)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
#plt.yscale('log') #logでスケールさせたいとき
plt.savefig(f'{homePath}/Scaling/TTS099_scale_var_tts.pdf', bbox_inches='tight') #名前変える
plt.show()
```

```python
# サブプロットのレイアウトを設定
fig, axes = plt.subplots(ncols=3, figsize=(12, 10), sharey=True)
fig.subplots_adjust(hspace=0.5)


QA_TTSs = [QA_TTS, QA_TTS0999, QA_TTS09999]
SQA_TTSs = [SQA_TTS, SQA_TTS0999, SQA_TTS09999]
titles = ["r_m=0.99", "r_m=0.999", "r_m=0.9999"]


# 各条件と折れ線グラフを配置
for i in range(3):
    ax = axes[i]

    # 各条件ごとの折れ線グラフを追加
    ax.plot(transactions, QA_TTSs[i], label="QA", color="darkorange")
    ax.plot(transactions, SQA_TTSs[i], label="SQA", color="blue")
    ax.plot(transactions, SMT_times, label="SMT", color="Red")

    ax.set_title(titles[i])
    ax.legend()

# グラフの共通のラベルとタイトルを設定
fig.text(0.5, 0.04, 'X-axis', ha='center')
fig.text(0.04, 0.5, 'Y-axis', va='center', rotation='vertical')
fig.suptitle("Multiple Line Charts with Conditions", fontsize=16)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
conditions = ['Condition 1', 'Condition 2', 'Condition 3']
colors = ['blue', 'orange', 'green']
linestyles = ['solid', 'dashed', 'dotted']
markers = ['o', 's', '^']

# サブプロットのレイアウトを設定
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5), sharex=True, sharey=True)
fig.subplots_adjust(wspace=0.25)

# 各Conditionごとにサブプロットに折れ線グラフを追加
for i, condition in enumerate(conditions):
    ax = axes[i]

    for j in range(3):
        y = [10, 12, 8, 5, 14]  # 仮のデータ

        # 同じCondition内で異なるラインスタイルとマーカーの折れ線グラフを追加
        ax.plot(x, y, label=f"{linestyles[j]} - {markers[j]}", color=colors[i], linestyle=linestyles[j], marker=markers[j])

    ax.set_title(condition)
    ax.legend()

# グラフの共通のラベルとタイトルを設定
fig.text(0.5, 0.04, 'X-axis', ha='center')
fig.text(0.04, 0.5, 'Y-axis', va='center', rotation='vertical')
fig.suptitle("Multiple Line Charts for Different Conditions", fontsize=16)

plt.show()
```
