```python
%matplotlib inline
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import os 
import json
```

```python
path = os.getcwd()
print(path)

os.chdir('../../')
homePath = os.getcwd()
print(homePath)
```

```python
df_sqa = pd.read_json(f'{homePath}/MakeGraph/twitter10000/twitter10000.json') #ここ変更すればjson変えられる
display(df_sqa)

#SMTの実行時間を計測しておく
SMT_time = 318
```

```python
plt.figure(figsize=(5,4))

sqa_time = df_sqa['time']
sqa_fsr = df_sqa['feasibleRate']
plt.plot(sqa_time, sqa_fsr, label='SQA')

plt.axvline(x=SMT_time, color='red', linestyle='--', label = 'SMT')

#ラベルなどの出力
plt.legend(fontsize=13)
plt.xlabel("Execution time(ms)",fontsize=13)
plt.ylabel("Feasible rate",fontsize=13)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xscale('log') #logでスケールさせたいとき
plt.savefig(f'{homePath}/MakeGraph/twitter10000/twitter10000_time_fsr.pdf') #名前変える
plt.show()
```

```python
df_sqa = df_sqa.sort_values('TTS')
display(df_sqa)
```

```python
plt.figure(figsize=(5,4))

sqa_tts = df_sqa['TTS']
plt.plot(sqa_tts, sqa_fsr, label='SQA')

plt.axvline(x=SMT_time, color='red', linestyle='--', label = 'SMT')

#ラベルなどの出力
plt.legend(fontsize=13)
plt.xlabel("TTS(ms)",fontsize=13)
plt.ylabel("Feasible rate",fontsize=13)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
#plt.xscale('log') #logでスケールさせたいとき
plt.savefig(f'{homePath}/MakeGraph/twitter10000/twitter10000_tts_fsr.pdf') #名前変える
plt.show()
```
