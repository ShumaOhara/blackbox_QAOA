```python
import pandas as pd
import os 
import json
import math
```

```python
def TimeToSolution(tau, feasibleRate, targetProbability):
    time = tau.copy()  # 入力のコピーを作成
    
    # シリーズ内の各要素に対して処理
    for i in range(len(tau)):
        if feasibleRate.iloc[i] == 1:  # インデックスiの要素にアクセス
            time.iloc[i] = tau.iloc[i]
        else:
            time.iloc[i] = tau.iloc[i] * math.ceil(math.log(1 - targetProbability) / math.log(1 - feasibleRate.iloc[i]))
    
    return time
```

```python
def Edit_TTS(filePath):
    #input
    df = pd.read_json(filePath) #ここ変更すればjson変えられる
    print("dataframe before edit")
    display(df)
    
    #edit part
    df['TTS(0999)'] = TimeToSolution(df['time'], df['feasibleRate'], 0.999)
    df['TTS(09999)'] = TimeToSolution(df['time'], df['feasibleRate'], 0.9999)
    display(df)
    
    #output
    df.to_json(filePath)
```

```python
#パスをコピーすると楽
filePath = 'twitter10000/SQAtwitter10000.json'
Edit_TTS(filePath)
```
